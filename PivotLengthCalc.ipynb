# PIVOT LENGTH ANALYSIS


```python
import numpy as np 
import pandas as pd 
import math
```

```python
# MEASUREMENTS 

DELTA_Z = 2.1033
TOOL_LENGTH = 10.2307
TOOL_OFFSET = .5
THETA = 30
```

```python
math.cos(30/180 * math.pi)
```

```python
def pivotLength(delta_Z, gauge_Offset, theta, tool_Length):

    theta_rad = math.cos(theta/180 *math.pi)

    return ((delta_Z + gauge_Offset*(1-theta_rad))/(1-theta_rad)) - tool_Length
```

```python
print(pivotLength(DELTA_Z,TOOL_OFFSET,THETA,TOOL_LENGTH))
```

```python
print(pivotLength(1.4368,0,30,4.75))
```
